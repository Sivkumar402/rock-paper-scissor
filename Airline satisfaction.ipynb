Box Plot

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
columns_to_plot = ['Departure/Arrival time convenient', 'On-board service', 'Leg room service', 'Cleanliness']
melted_data = df[columns_to_plot].melt(var_name='Columns', value_name='Values')
plt.figure(figsize=(12, 6))
sns.boxplot(x='Columns', y='Values', data=melted_data)
plt.title('Boxplot')
plt.xlabel('Columns')
plt.ylabel('Values')
plt.show()
```
